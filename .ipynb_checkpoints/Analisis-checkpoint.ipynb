### Convertimos los datos en matrices y vectores para realizar análisis de $\chi ^2$

```python
archivoArray = np.array(archivo)
archivoMat = []
for i in archivoArray:
    archivoMat.append(i[1:])
archivoMat = np.array(archivoMat)
archivoMat
```

```python
total
```

```python
totalTema
```

# Prueba $\chi ^2$:
## $H_0$ : Los colores no influyen en los temas.
## $H_1$ : Los colores sí influyen en los temas.

### Obtenemos las frecuencias teóricas.

```python
#Total 
tot = np.array(total).sum()
# Frecuenias teóricas.
ft = []
contTema = 0
for i in archivoMat:
    contCol = 0
    for j in i:
        ft.append(total[contCol]*totalTema[contTema]/tot)
        contCol += 1
    contTema += 1
ft = np.array(ft)
ft
```

### Grados de libertad.

```python
#Grado de libertad
dim = archivoMat.shape
gd = (dim[0]-1)*(dim[1]-1)
gd
```

### $\chi ^ 2$

```python
#Chi cuadrado
archivoMatArray = []
for i in archivoMat:
    for j in i:
        archivoMatArray.append(j)

chi = 0
for i in range(archivoMat.size):
        chi += (archivoMatArray[i] - ft[i])**2 / ft[i]
chi
```

### Comparamos $\chi ^2 _{calculada}$ con los valores en tablas.

```python
print("Grados de libertad: " + str(gd), "; chi cuadrada: " + str(chi) )
```
